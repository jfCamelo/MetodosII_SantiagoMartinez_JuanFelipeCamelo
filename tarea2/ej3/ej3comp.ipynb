```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def f(x,y):
    return x*y**2-2*x**-1*y-x**-3
```

```python
def Solucion(x):
    return 1/(x**2+2)*(2/x**2-1)
```

```python
def Solucion(x):
    return (x**-2+1/2)/(1/2*x**2-1)
```

```python
def GetRungeKutta4(x,y,h):
    for i in range(1,len(x)):
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h/2,y[i-1]+h/2*k1)
        k3 = f(x[i-1]+h/2,y[i-1]+h/2*k2)
        k4 = f(x[i-1]+h,y[i-1]+h*k3)

        y[i] = y[i-1] + h/6*(k1+2*k2+2*k3+k4)
```

```python
N = 200
x = np.arange(np.sqrt(2), 2, 0.1)
h = 0.1
y4 = np.zeros_like(x)
y4[0] = 0.
GetRungeKutta4(x,y4,h)
yext = Solucion(x)
```

```python
plt.plot(x,y4,label='RungeKutta4')
plt.plot(x,yext,label='Solucion')
plt.legend()
plt.title("Comparación de la solución exacta y la solución numérica con RungeKutta4")
```
