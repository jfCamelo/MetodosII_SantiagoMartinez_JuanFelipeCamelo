```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def f(x,y,q):
    return y**q
```

```python
N = 10
t = np.linspace(0,1,N)
```

```python
h = t[1] - t[0]
```

```python
def GetRungeKutta3(f, x, y, h, q):
    for i in range(1, len(x)):
        k1 = f(x[i-1], y[i-1], q)
        k2 = f(x[i-1] + h/2, y[i-1] + h/2*k1, q)
        k3 = f(x[i-1] + h, y[i-1] - h*k1 + 2*h*k2, q)
        y[i] = y[i-1] + h/6 * (k1 + 4*k2 + k3)
```

```python
y1 = np.zeros_like(t)
y1[0] = 1
```

```python
q = 0
GetRungeKutta3(f,t,y1,h,q)
plt.plot(t,y1, label='q = 0')

q = 0.2
GetRungeKutta3(f,t,y1,h,q)
plt.plot(t,y1, label='q = 0.2')

q = 0.4
GetRungeKutta3(f,t,y1,h,q)
plt.plot(t,y1, label='q = 0.4')

q = 0.7
GetRungeKutta3(f,t,y1,h,q)
plt.plot(t,y1, label='q = 0.7')

q = 0.9
GetRungeKutta3(f,t,y1,h,q)
plt.plot(t,y1, label='q = 0.9')

q = 1
GetRungeKutta3(f,t,y1,h,q)
plt.plot(t,y1, label='q = 1')

plt.legend()
```

Ahora podemos comprobar con nuestros resultados del anterior punto

```python
def f1(x,y):
    return np.exp(x)
```

```python
def f2(x,y,q):
    return (x*(1-q)+1)**(1/(1-q))
```

```python
plt.plot(t, f2(t,0,0), label='q=0')
plt.plot(t, f2(t,0,0.2), label='q=0.2')
plt.plot(t, f2(t,0,0.4), label='q=0.4')
plt.plot(t, f2(t,0,0.7), label='q=0.7')
plt.plot(t, f2(t,0,0.9), label='q=0.9')
plt.plot(t, f1(t,0), label='q=1')
plt.legend()
```
